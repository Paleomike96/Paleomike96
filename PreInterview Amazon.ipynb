Aquí vamos a comenzar con nuestra prueba de estructura de datos

```python
def sortInWave(arr):
    n = len(arr)

    # Traverse all even index elements
    for i in range(0, n, 2):

        # If current even element is smaller than the previous element
        if i > 0 and arr[i] < arr[i - 1]:
            arr[i], arr[i - 1] = arr[i - 1], arr[i]

        # If current even element is smaller than the next element
        if i < n - 1 and arr[i] < arr[i + 1]:
            arr[i], arr[i + 1] = arr[i + 1], arr[i]

# Example usage
arr = [1, 2, 3, 4, 5, 6,7,8]
sortInWave(arr)
print(arr)
```
